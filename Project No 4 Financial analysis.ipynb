```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('C:/Users/PRUTHVI/Desktop/COLLEGE/University/Project Work/UnifiedMentor/Financial Analytics data.csv')
```

```python
print(data.head())
```

```python
print(data.describe())
```

```python
total_market_cap = data['Mar Cap - Crore'].sum()
total_sales = data['Sales Qtr - Crore'].sum()
print('Total Market Cap:',total_market_cap)
print('Total Quarterly Sales:',total_sales)
```

```python
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
sns.histplot(data['Mar Cap - Crore'], bins=30, kde=True)
plt.title('Market Capitalization Distribution')
```

```python
plt.subplot(1, 1, 1)
sns.histplot(data['Sales Qtr - Crore'], bins=30, kde=True)
plt.title('Quarterly Sales Distribution')
plt.show()
```

```python
# Calculate correlation
correlation = data['Mar Cap - Crore'].corr(data['Sales Qtr - Crore'])
print(f'Correlation between Market Capitalization and Quarterly Sales: {correlation}')
```

```python
# Scatter plot to show the relationship
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Sales Qtr - Crore', y='Mar Cap - Crore', data=data)
plt.title('Market Capitalization vs Quarterly Sales')
plt.xlabel('Quarterly Sale in crores')
plt.ylabel('Market Capitalization in Crores')
plt.show()
```

```python
top_10_companies = data.nlargest(10, 'Mar Cap - Crore')
print(top_10_companies[['Name', 'Mar Cap - Crore']])
```

```python
top_10_companies_by_quarterly_sales = data.nlargest(10, 'Sales Qtr - Crore')
print(top_10_companies_by_quarterly_sales[['Name', 'Sales Qtr - Crore']])
```

```python
# Sales Efficiency
data['Sales Efficiency'] = data['Sales Qtr - Crore'] / data['Mar Cap - Crore']
print(data[['Name', 'Sales Efficiency']])
```

```python
top_10_companies_by_sales_efficiency = data.nlargest(10, 'Sales Efficiency')
print(top_10_companies_by_sales_efficiency[['Name', 'Sales Efficiency']])
```
